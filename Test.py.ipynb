```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## Using the yfinance Library to Extract Stock Data


Using the `Ticker` module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is `AAPL`.


```python
apple = yf.Ticker("AAPL")
```

Now we can access functions and variables to extract the type of data we need. You can view them and what they represent here https://aroussi.com/post/python-yahoo-finance.


```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json
```

### Stock Info


Using the attribute  <code>info</code> we can extract information about the stock as a Python dictionary.


```python
import json
with open('apple.json') as json_file:
    apple_info = json.load(json_file)
    # Print the type of data variable    
    #print("Type:", type(apple_info))
apple_info
```

#### APPLE country


```python
apple_info['country']
```

### Extracting Share Price


#### apple share price



```python
apple_share_price_data = apple.history(period="max")
```

The format that the data is returned in is a Pandas


```python
apple_share_price_data.head()
```

### Visualization


```python
apple_share_price_data.plot(x="Date", y="Open")
```

#### Apple stock prices


```python
plt.plot(apple_share_price_data["Date"], apple_share_price_data["Close"], label="Apple")
# Set x-axis label and title
plt.xlabel("Date")
plt.title(" Apple Stock Prices")
```


## Using the yfinance Library to Extract Stock Data



Using the `Ticker` module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is 'AMD'

```python
amd =yf.Ticker("AMD")
amd
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/amd.json
```

```python
import json
with open('amd.json') as json_file:
    amd_info = json.load(json_file)
    # Print the type of data variable    
    #print("Type:", type(amd_info))
amd_info
```

#### AMD COUNTRY


```python
amd_info['country']
```

### Extracting Share Price

```python
amd_stock_data = amd.history(period='max')
```

```python
amd_stock_data.head()
```

### Visualization


```python
amd_stock_data.plot(x="Date", y="Open")
```


#### AMD stock prices


```python
plt.plot(amd_stock_data["Date"], amd_stock_data["Close"], label="AMD")
# Set x-axis label and title
plt.xlabel("Date")
plt.title("AMD")
```

## AMD vs Apple Stock Prices




```python


# Plot AMD stock price
plt.plot(amd_stock_data["Date"], amd_stock_data["Close"], label="AMD")

# Plot Apple stock price
plt.plot(apple_share_price_data["Date"], apple_share_price_data["Close"], label="Apple")

# Set x-axis label and title
plt.xlabel("Date")
plt.title("AMD vs. Apple Stock Prices")

# Show legend
plt.legend()

# Show the plot
plt.show()
```
